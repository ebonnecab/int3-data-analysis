**Note**:My data is definitely skewed so I'm going to work on getting rid of the outliers for a more accurate analysis.

```python
import numpy as np

def find_remove_outlier(data_sample):
    data_mean = sum(data_sample)/len(data_sample)
    data_std = np.std(data_sample)
    
    cut_off = data_std * 3
    
    lower, upper = data_mean - cut_off, data_mean + cut_off
    
    outliers = []
    removed_outliers = []
    
    for x in data_sample:
        if x < lower or x > upper:
            outliers.append(x)
        if x > lower and x < upper:
            removed_outliers.append(x)
    return removed_outliers   
```

```python
print(find_remove_outlier(df['suicides_no']))
```
